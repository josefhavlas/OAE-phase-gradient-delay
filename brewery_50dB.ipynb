CALCULATING NEEDED VALUES

SFOAE

```python
# peak picking
maximas_SFOAE = findLocalMaxima(np.real(20*np.log10(SFOAE)))
shift = 1  # 3 points around all peaks (the peak is included)
ext_maximas_SFOAE = peakNeighbours(maximas_SFOAE, shift, freq_range_SFOAE)

# unwrapped phase of SFOAE and its gradient ~ phase delay
phase_unwrap_SFOAE = np.unwrap(np.angle(SFOAE))
# difference between neighbour samples on the frequency axis (evenly spaced)
# step_size_SFOAE = freq_SFOAE[1] - freq_SFOAE[0]
# phase_grad_SFOAE = np.gradient(phase_unwrap_SFOAE, step_size_SFOAE)
phase_grad_SFOAE = np.gradient(phase_unwrap_SFOAE, freq_SFOAE)
group_delay_SFOAE = -phase_grad_SFOAE/(2*np.pi)
```

CR COMPONENT

```python
freq_range_CR = freq_range_SFOAE
freq_CR = freq_SFOAE

# peak picking
maximas_CR = findLocalMaxima(np.real(20*np.log10(CRcomp)))
shift = 1  # 3 points around all peaks (the peak is included)
ext_maximas_CR = peakNeighbours(maximas_CR, shift, freq_range_CR)

# unwrapped phase of CR COMPONENT and its gradient ~ phase delay
phase_unwrap_CR = np.unwrap(np.angle(CRcomp))
# difference between neighbour samples on the frequency axis (evenly spaced)
# step_size_CR = step_size_SFOAE
# phase_grad_CR = np.gradient(phase_unwrap_CR, step_size_CR)
phase_grad_CR = np.gradient(phase_unwrap_CR, freq_CR)
group_delay_CR = -phase_grad_CR/(2*np.pi)
```

NL COMPONENT

```python
freq_range_NL = freq_range_SFOAE
freq_NL = freq_SFOAE

# peak picking
maximas_NL = findLocalMaxima(np.real(20*np.log10(NLcomp)))
shift = 1  # 3 points around all peaks (the peak is included)
ext_maximas_NL = peakNeighbours(maximas_NL, shift, freq_range_NL)

# unwrapped phase of NL COMPONENT and its gradient ~ phase delay
phase_unwrap_NL = np.unwrap(np.angle(NLcomp))
# difference between neighbour samples on the frequency axis (evenly spaced)
# step_size_NL = step_size_SFOAE
# phase_grad_NL = np.gradient(phase_unwrap_NL, step_size_NL)
phase_grad_NL = np.gradient(phase_unwrap_NL, freq_NL)
group_delay_NL = -phase_grad_NL/(2*np.pi)
```
