DATA LOADING

```python
"""
Different inhomogeneities of SFOAE 20 dB SPL data

Author: havlajos
Created: 20/05/2023
Last Update: 11/07/2023

Description: Analysis of different SFOAE inhimogeneities used in the nonlinear cochlear model.
The SFOAE data is of 20 dB SPL. Phase-gradient delays and amplitudes are compared.
"""

import numpy as np
import matplotlib.pyplot as plt
import scipy.io

# script importing
from peak_picking import *
from gradient_delay import *

# SFOAE data loading
mat_SFOAE_19 = scipy.io.loadmat('data/SFOAE_20dB_inhom/sfoae20dBg130TM45R19rs2.mat')

freq_SFOAE_19 = mat_SFOAE_19['Fvect'].flatten()  # frequency axis
SFOAE_19 = mat_SFOAE_19['oaeNum20'].flatten()  # SFOAE wave
CRcomp_19 = mat_SFOAE_19['CRc20'].flatten()  # CR component
NLcomp_19 = mat_SFOAE_19['Yunl20'].flatten()  # NL component

mat_SFOAE_23 = scipy.io.loadmat('data/SFOAE_20dB_inhom/sfoaeR23rs20.mat')

freq_SFOAE_23 = mat_SFOAE_23['Fvect'].flatten()  # frequency axis
SFOAE_23 = mat_SFOAE_23['SFOAEan20'].flatten()  # SFOAE wave
CRcomp_23 = mat_SFOAE_23['CRc20'].flatten()  # CR component
NLcomp_23 = mat_SFOAE_23['Yunl20'].flatten()  # NL component

mat_SFOAE_24 = scipy.io.loadmat('data/SFOAE_20dB_inhom/sfoaeR24rs20.mat')

freq_SFOAE_24 = mat_SFOAE_24['Fvect'].flatten()  # frequency axis
SFOAE_24 = mat_SFOAE_24['SFOAEan20'].flatten()  # SFOAE wave
CRcomp_24 = mat_SFOAE_24['CRc20'].flatten()  # CR component
NLcomp_24 = mat_SFOAE_24['Yunl20'].flatten()  # NL component

mat_SFOAE_25 = scipy.io.loadmat('data/SFOAE_20dB_inhom/sfoaeR25rs20.mat')

freq_SFOAE_25 = mat_SFOAE_25['Fvect'].flatten()  # frequency axis
SFOAE_25 = mat_SFOAE_25['SFOAEan20'].flatten()  # SFOAE wave
CRcomp_25 = mat_SFOAE_25['CRc20'].flatten()  # CR component
NLcomp_25 = mat_SFOAE_25['Yunl20'].flatten()  # NL component

freq_range_19 = np.size(freq_SFOAE_19)  # maximum index on the frequency axis
freq_range_23 = np.size(freq_SFOAE_23)
freq_range_24 = np.size(freq_SFOAE_24)
freq_range_25 = np.size(freq_SFOAE_25)
```

CALCULATING NEEDED VALUES

SFOAE

```python
# peak picking
maximas_SFOAE_19 = findLocalMaxima(np.real(20*np.log10(SFOAE_19)))
shift = 1  # 3 points around all peaks (the peak is included)
ext_maximas_SFOAE_19 = peakNeighbours(maximas_SFOAE_19, shift, freq_range_19)
group_delay_SFOAE_19 = calcDelay(freq_SFOAE_19, SFOAE_19)

maximas_SFOAE_23 = findLocalMaxima(np.real(20*np.log10(SFOAE_23)))
ext_maximas_SFOAE_23 = peakNeighbours(maximas_SFOAE_23, shift, freq_range_23)
group_delay_SFOAE_23 = calcDelay(freq_SFOAE_23, SFOAE_23)

maximas_SFOAE_24 = findLocalMaxima(np.real(20*np.log10(SFOAE_24)))
ext_maximas_SFOAE_24 = peakNeighbours(maximas_SFOAE_24, shift, freq_range_24)
group_delay_SFOAE_24 = calcDelay(freq_SFOAE_24, SFOAE_24)

maximas_SFOAE_25 = findLocalMaxima(np.real(20*np.log10(SFOAE_25)))
ext_maximas_SFOAE_25 = peakNeighbours(maximas_SFOAE_25, shift, freq_range_25)
group_delay_SFOAE_25 = calcDelay(freq_SFOAE_25, SFOAE_25)
```

CR COMPONENT

```python
# peak picking
maximas_CR_19 = findLocalMaxima(np.real(20*np.log10(CRcomp_19)))
shift = 1  # 3 points around all peaks (the peak is included)
ext_maximas_CR_19 = peakNeighbours(maximas_CR_19, shift, freq_range_19)
group_delay_CR_19 = calcDelay(freq_SFOAE_19, CRcomp_19)

maximas_CR_23 = findLocalMaxima(np.real(20*np.log10(CRcomp_23)))
ext_maximas_CR_23 = peakNeighbours(maximas_CR_23, shift, freq_range_23)
group_delay_CR_23 = calcDelay(freq_SFOAE_23, CRcomp_23)

maximas_CR_24 = findLocalMaxima(np.real(20*np.log10(CRcomp_24)))
ext_maximas_CR_24 = peakNeighbours(maximas_CR_24, shift, freq_range_24)
group_delay_CR_24 = calcDelay(freq_SFOAE_24, CRcomp_24)

maximas_CR_25 = findLocalMaxima(np.real(20*np.log10(CRcomp_25)))
ext_maximas_CR_25 = peakNeighbours(maximas_CR_25, shift, freq_range_25)
group_delay_CR_25 = calcDelay(freq_SFOAE_25, CRcomp_25)
```

NL COMPONENT

```python

# peak picking
maximas_NL_19 = findLocalMaxima(np.real(20*np.log10(NLcomp_19)))
ext_maximas_NL_19 = peakNeighbours(maximas_NL_19, shift, freq_range_19)
group_delay_NL_19 = calcDelay(freq_SFOAE_19, NLcomp_19)

maximas_NL_23 = findLocalMaxima(np.real(20*np.log10(NLcomp_23)))
ext_maximas_NL_23 = peakNeighbours(maximas_NL_23, shift, freq_range_23)
group_delay_NL_23 = calcDelay(freq_SFOAE_23, NLcomp_23)

maximas_NL_24 = findLocalMaxima(np.real(20*np.log10(NLcomp_24)))
ext_maximas_NL_24 = peakNeighbours(maximas_NL_24, shift, freq_range_24)
group_delay_NL_24 = calcDelay(freq_SFOAE_24, NLcomp_24)

maximas_NL_25 = findLocalMaxima(np.real(20*np.log10(NLcomp_25)))
ext_maximas_NL_25 = peakNeighbours(maximas_NL_25, shift, freq_range_25)
group_delay_NL_25 = calcDelay(freq_SFOAE_25, NLcomp_25)
```

ISOLATED INTENSITY

```python
from isoint import *

ISOINT_start, ISOINT_stop, ISOINT_step = 1000, 3200, 200

freq_ISOINT_list = []
GD_ISOINT_list = []
max_idx_ISOINT_list = []

for i in range(ISOINT_start, ISOINT_stop + 1, ISOINT_step):
    filename = f"data/isointFT/isoint_{i}Hz_Level_20dB_TM45130.mat"
    freq_ISOINT, group_delay_ISOINT, max_idx_ISOINT = processIsoInt(filename)

    freq_ISOINT_list.append(freq_ISOINT)
    GD_ISOINT_list.append(group_delay_ISOINT)
    max_idx_ISOINT_list.append(max_idx_ISOINT)
```

PLOTTING

```python
from matplotlib.ticker import FixedLocator, FixedFormatter

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(24, 14))
fig.supxlabel('Frequency (kHz)', fontsize=30)
fig.supylabel('Phase-gradient delay (ms)', fontsize=30)
fig.subplots_adjust(left=0.06, bottom=0.07)

# multiplying by 1000 because of the transfer to miliseconds
ax1.plot(freq_SFOAE_19/1000, 1000*group_delay_SFOAE_19, '0.8', linestyle='--', label='SFOAE delay')

# reference values
for i, freq_ISOINT in enumerate(freq_ISOINT_list):
    max_idx = max_idx_ISOINT_list[i]
    grad_delay_ISOINT = GD_ISOINT_list[i]
    ax1.plot(freq_ISOINT[max_idx]/1000, 1000*grad_delay_ISOINT[max_idx],
            "bD", markersize=12, label='reference' if i == 0 else None)
    if i > 0:  # connect points starting from the second point
        ax1.plot([freq_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]]/1000, freq_ISOINT[max_idx]/1000],
                [1000*GD_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]],
                    1000*grad_delay_ISOINT[max_idx]],
                "b--")

ax1.plot(freq_SFOAE_19[ext_maximas_CR_19]/1000, 1000*group_delay_CR_19[ext_maximas_CR_19],
        "g^", markersize=12, label='peak-picking CR')

ax1.plot(freq_SFOAE_19[ext_maximas_NL_19]/1000, 1000*group_delay_NL_19[ext_maximas_NL_19],
        "kv", markersize=12, label='peak-picking NL')

ax1.plot(freq_SFOAE_19[ext_maximas_SFOAE_19]/1000, 1000*group_delay_SFOAE_19[ext_maximas_SFOAE_19],
        "rx", markersize=12, label='peak-picking SFOAE')

ax1.set_xscale('log')
ax1.set_xlim([1, 3.5])
# Customize the tick labels on the x-axis
x_ticks = [1, 1.5, 2, 2.5, 3, 3.5]
ax1.xaxis.set_major_locator(FixedLocator(x_ticks))
ax1.xaxis.set_major_formatter(FixedFormatter(x_ticks))
ax1.tick_params(axis='both', which='major', labelsize=25)

ax1.set_ylim([2, 20])

ax1.legend(fontsize=16)

# multiplying by 1000 because of the transfer to miliseconds
ax2.plot(freq_SFOAE_23/1000, 1000*group_delay_SFOAE_23,
         '0.8', linestyle='--', label='group delay SFOAE')

# reference values
for i, freq_ISOINT in enumerate(freq_ISOINT_list):
    max_idx = max_idx_ISOINT_list[i]
    grad_delay_ISOINT = GD_ISOINT_list[i]
    ax2.plot(freq_ISOINT[max_idx]/1000, 1000*grad_delay_ISOINT[max_idx],
             "bD", markersize=12, label='reference' if i == 0 else None)
    if i > 0:  # connect points starting from the second point
        ax2.plot([freq_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]]/1000, freq_ISOINT[max_idx]/1000],
                 [1000*GD_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]],
                  1000*grad_delay_ISOINT[max_idx]],
                 "b--")

ax2.plot(freq_SFOAE_23[ext_maximas_CR_23]/1000, 1000*group_delay_CR_23[ext_maximas_CR_23],
         "g^", markersize=12, label='peak picking CR')

ax2.plot(freq_SFOAE_23[ext_maximas_NL_23]/1000, 1000*group_delay_NL_23[ext_maximas_NL_23],
         "kv", markersize=12, label='peak picking NL')

ax2.plot(freq_SFOAE_23[ext_maximas_SFOAE_23]/1000, 1000*group_delay_SFOAE_23[ext_maximas_SFOAE_23],
         "rx", markersize=12, label='peak picking SFOAE')

ax2.set_xscale('log')
ax2.set_xlim([1, 3.5])
# Customize the tick labels on the x-axis
x_ticks = [1, 1.5, 2, 2.5, 3, 3.5]
ax2.xaxis.set_major_locator(FixedLocator(x_ticks))
ax2.xaxis.set_major_formatter(FixedFormatter(x_ticks))
ax2.tick_params(axis='both', which='major', labelsize=25)

ax2.set_ylim([2, 20])

# multiplying by 1000 because of the transfer to miliseconds
ax3.plot(freq_SFOAE_24/1000, 1000*group_delay_SFOAE_24,
         '0.8', linestyle='--', label='group delay SFOAE')

# reference values
for i, freq_ISOINT in enumerate(freq_ISOINT_list):
    max_idx = max_idx_ISOINT_list[i]
    grad_delay_ISOINT = GD_ISOINT_list[i]
    ax3.plot(freq_ISOINT[max_idx]/1000, 1000*grad_delay_ISOINT[max_idx],
             "bD", markersize=12, label='reference' if i == 0 else None)
    if i > 0:  # connect points starting from the second point
        ax3.plot([freq_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]]/1000, freq_ISOINT[max_idx]/1000],
                 [1000*GD_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]],
                  1000*grad_delay_ISOINT[max_idx]],
                 "b--")

ax3.plot(freq_SFOAE_24[ext_maximas_CR_24]/1000, 1000*group_delay_CR_24[ext_maximas_CR_24],
         "g^", markersize=12, label='peak picking CR')

ax3.plot(freq_SFOAE_24[ext_maximas_NL_24]/1000, 1000*group_delay_NL_24[ext_maximas_NL_24],
         "kv", markersize=12, label='peak picking NL')

ax3.plot(freq_SFOAE_24[ext_maximas_SFOAE_24]/1000, 1000*group_delay_SFOAE_24[ext_maximas_SFOAE_24],
         "rx", markersize=12, label='peak picking SFOAE')

ax3.set_xscale('log')
ax3.set_xlim([1, 3.5])
# Customize the tick labels on the x-axis
x_ticks = [1, 1.5, 2, 2.5, 3, 3.5]
ax3.xaxis.set_major_locator(FixedLocator(x_ticks))
ax3.xaxis.set_major_formatter(FixedFormatter(x_ticks))
ax3.tick_params(axis='both', which='major', labelsize=25)

ax3.set_ylim([2, 20])

# multiplying by 1000 because of the transfer to miliseconds
ax4.plot(freq_SFOAE_25/1000, 1000*group_delay_SFOAE_25,
         '0.8', linestyle='--', label='group delay SFOAE')

# reference values
for i, freq_ISOINT in enumerate(freq_ISOINT_list):
    max_idx = max_idx_ISOINT_list[i]
    grad_delay_ISOINT = GD_ISOINT_list[i]
    ax4.plot(freq_ISOINT[max_idx]/1000, 1000*grad_delay_ISOINT[max_idx],
             "bD", markersize=12, label='reference' if i == 0 else None)
    if i > 0:  # connect points starting from the second point
        ax4.plot([freq_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]]/1000, freq_ISOINT[max_idx]/1000],
                 [1000*GD_ISOINT_list[i-1][max_idx_ISOINT_list[i-1]],
                  1000*grad_delay_ISOINT[max_idx]],
                 "b--")

ax4.plot(freq_SFOAE_25[ext_maximas_CR_25]/1000, 1000*group_delay_CR_25[ext_maximas_CR_25],
         "g^", markersize=12, label='peak picking CR')

ax4.plot(freq_SFOAE_25[ext_maximas_NL_25]/1000, 1000*group_delay_NL_25[ext_maximas_NL_25],
         "kv", markersize=12, label='peak picking NL')

ax4.plot(freq_SFOAE_25[ext_maximas_SFOAE_25]/1000, 1000*group_delay_SFOAE_25[ext_maximas_SFOAE_25],
         "rx", markersize=12, label='peak picking SFOAE')

ax4.set_xscale('log')
ax4.set_xlim([1, 3.5])
# Customize the tick labels on the x-axis
x_ticks = [1, 1.5, 2, 2.5, 3, 3.5]
ax4.xaxis.set_major_locator(FixedLocator(x_ticks))
ax4.xaxis.set_major_formatter(FixedFormatter(x_ticks))
ax4.tick_params(axis='both', which='major', labelsize=25)

ax4.set_ylim([2, 20])

plt.rcParams['font.family'] = 'Times New Roman'
# plt.savefig('sfoae_inhom_PP_all.png', dpi=300)
plt.show()

```
